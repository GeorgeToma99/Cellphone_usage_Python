```python
import pandas as pd  
import numpy as np
import matplotlib.pyplot as plt  
import seaborn as sns
```

```python
Phone_df = pd.read_csv('../../data/mobile_usage_behavioral_analysis.csv')
```

```python
Phone_df.head()
```

```python
Gender_df = Phone_df.groupby('Gender').agg({'Daily_Screen_Time_Hours': 'sum'})
```

```python
print(Gender_df)
```

```python
Age_df = Phone_df.groupby('Age').agg({'Daily_Screen_Time_Hours': 'sum'}).reset_index()
```

```python
Age_df.head()
```

```python
Age_df.sort_values(by = 'Daily_Screen_Time_Hours', ascending=False)
```

```python
Age_df['Decade'] = (Age_df['Age'] // 10) * 10
```

```python
Age_df.head()
```

```python
Decade_Age_df = Age_df.groupby('Decade').agg({'Daily_Screen_Time_Hours': 'sum'}).reset_index()
```

```python
Decade_Age_df.sort_values(by = 'Daily_Screen_Time_Hours', ascending=False)
```
